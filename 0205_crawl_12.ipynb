### request - 요청
* 우리가 주소창에 naver.com 입력해서 네이버 홈페이지로 들어갔다는 것은
* naver.com 내용으로 요청(request)을 한 결과가 내 컴퓨터로 왔고
* 그 결과를 구글 크롬과 같은 웹 브라우저가 해석해보니 네이버 홈페이지인 것입니다.
* 스마트폰, 컴퓨터 등 알게 모르게 요청을 사용하고 있었습니다.

```python
import requests # 요청을 하는 도구를 가져옵니다.
# port ~> 항구, im ~> 안으로~ 라는 의미. 
# import -> 결국 우리나라에 없는 물건을 해외에서 들여오는 것(수입)
# 가져올 것이다~ 라는 의미입니다.
url = 'https://api.sunrise-sunset.org/json?lat=36.7201600&lng=-4.4203400'
data = requests.get(url).json()
data
# 요청에는 종류가 있는데, 나에게 정보를 달라(get)는 것과 내 정보를 받아(post)달라.
# 우리는 정보를 가져오는 것이 주 목적이므로, get만 사용합니다.
# 딕셔너리처럼 생긴 자료의 형태를 웹 동네에서는 JSON 이라고 합니다.
# .json 메서드는 requests 도구 만든 분들이, JSON 데이터를 쉽게 파이썬에서
# 사용하기 용이하게 JSON 형태를 파이썬식으로(딕셔너리로) 바꿔주는 것입니다.
```

##### by_date 함수, 무엇이 문제일까?
* 만약 10년치를 모아서 저장하면, 3650 일짜리 정보가 저장이 될 것입니다.
* 저장을 해도, 나중에 꺼내 쓸 때, 이 데이터가 몇일의 것인지를 특정지을 수 없다.
* 데이터베이스 기준으로는 Primary Key

```python
import requests
def by_date_2(date):
    url = 'https://api.sunrise-sunset.org/json?lat=37.5&lng=127.0&date='
    data = requests.get(url + date).json()['results']
    data['date'] = date # 날짜 정보가 키 : 값 쌍으로 추가되었기 때문에
    # 이제는 각 데이터가 몇일의 데이터인지 알 수 있겠습니다.
    return data
```

```python
import pandas as pd # pandas 도구를 가져와 pd 라는 축약어로 사용한다.
sample_list = []
sample_list.append(by_date_2('2020-01-01'))
sample_list.append(by_date_2('2020-01-02'))
sample_list.append(by_date_2('2020-01-03'))
df = pd.DataFrame(sample_list)
# pd.DataFrame 메서드는 데이터 프레임을 만들어주는 메서드입니다.
# 데이터 프레임이라는 형태로 만든 결과를 df 변수에 넣었습니다.
df
```
