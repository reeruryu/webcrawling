##### 새 친구 beautifulsoup 을 만나 봅니다.
* 같은 정보, 같은 로직, 같은 프로세스, 다른 도구(xmltodict -> bs4)

```python
import requests
import bs4
import pandas as pd
kw = '펭수'
url = f'https://news.google.com/rss/search?q={kw}&hl=ko&gl=KR&ceid=KR%3Ako'
data = requests.get(url).text
# before:
# news_data = xmltodict.parse(data)
news_data = bs4.BeautifulSoup(data, features='xml') # 파싱하는 메서드의 이름이 아름다운soup.
# features='xml' 부분은, 해석하는 해석기를 지정한 것입니다.
# before:
# news_list = news_data['rss']['channel']['item']
news_list = news_data.find_all('item') # 관심있는 키만 전부 찾기.
all_news_list = []
for news in news_list:
    # before:
    # title = news['title']
    # source = news['source']['#text']
    title = news.find('title').text # 태그는, XML 수업에서 처음에 키 라고 지칭 한 것.
    source = news.find('link').text # 뒤에 .text 적는 이유는, 태그 정보를 지우기 위해.
    one_news_dict = {'title' : title, 'source' : source}
    all_news_list.append(one_news_dict)
pd.DataFrame(all_news_list).to_csv(kw + '.csv', index=False, encoding='utf-8-sig')
```

```python
import requests
import pandas as pd
ing, menu = 'caramel', 'pizza'
url = f'http://www.recipepuppy.com/api/?i={ing}&q={menu}'
result = requests.get(url).json()
pd.DataFrame(result['results']).to_csv(f'{ing}_{menu}.csv', index=False)
```

```python
import requests
import pandas as pd
def recipe_get(ing, menu):
    url = f'http://www.recipepuppy.com/api/?i={ing}&q={menu}'
    result = requests.get(url).json()
    pd.DataFrame(result['results']).to_csv(f'{ing}_{menu}.csv', index=False)
recipe_get('chocolate', 'pizza')
```

```python
import xmltodict
ing, menu = 'caramel', 'pizza'
url = f'http://www.recipepuppy.com/api/?i={ing}&q={menu}&format=xml'
response = requests.get(url).text
result = xmltodict.parse(response)
pd.DataFrame(result['recipes']['recipe']).to_csv(f'{ing}_{menu}.csv', index=False)
```

```python
import bs4
ing, menu = 'caramel', 'pizza'
url = f'http://www.recipepuppy.com/api/?i={ing}&q={menu}&format=xml'
response = requests.get(url).text
# before : 
# result = xmltodict.parse(response)
result = bs4.BeautifulSoup(response)
title = [item.text for item in result.find_all('title')]
href = [item.text for item in result.find_all('href')]
ingredients = [item.text for item in result.find_all('ingredients')]
all_data = {'title' : title, 'href' : href, 'ingredients' : ingredients}
pd.DataFrame(all_data).to_csv(f'{ing}_{menu}.csv', index=False)
```
